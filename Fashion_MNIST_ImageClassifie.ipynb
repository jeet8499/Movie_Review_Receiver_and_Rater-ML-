```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
fashion_mnist = keras.datasets.fashion_mnist
(X_train_full,y_train_full),(X_test,y_test)= fashion_mnist.load_data()
```

```python
plt.imshow(X_train_full[1])
plt.show()
```

```python
y_train_full[2]
```

```python
class_name = ["T-shirt/top","Trouser","Pullover","Dress","Coat","Sandal","Shirt","Sneaker","Bag","Ankle boot"]
```

```python
class_name[y_train_full[0]]
```

```python
X_train_full[0]
```

```python
x_train_n =X_train_full / 255.0
x_test_n = X_test / 255.0
```

```python
x_valid, x_train = x_train_n[:5000],x_train_n[5000:]
y_valid , y_train = y_train_full[:5000],y_train_full[5000:]
x_test = x_test_n
```

```python
x_valid[0]
```

```python
np.random.seed(42)
tf.random.set_seed(42)
```

```python
model = keras.models.Sequential()
model.add(keras.layers.Flatten(input_shape=[28,28]))
model.add(keras.layers.Dense(300, activation="relu")) # ReLU (Rectified Linear Unit) activation: outputs input if positive, else zero. Introduces non-linearity.
model.add(keras.layers.Dense(100, activation="relu")) # ReLU activation: outputs input if positive, else zero.
model.add(keras.layers.Dense(10,activation="softmax")) # Softmax activation: used for multi-class classification output; converts scores to probabilities that sum to 1.
```

```python
model.summary()
```

```python
import pydot
keras.utils.plot_model(model)
```

```python
weights , biases = model.layers[1].get_weights()
```

```python
model.compile(loss = "sparsea_categorial_crossentropy",
              optimizer ="sgd",
              metrics = ["accuracy"])
```

```python
model_history = model.fit(x_train , y_train , epochs=30,
                          validation_data =(x_valid , y_valid))
```

```python
model_history.params
```

```python
import pandas as pd
pd.DataFrame(model_history.history).plot(figsize=(8,5))
plt.grid(True)
plt.gca().set_ylim(0,1)
plt.show()
```

```python
model.evaluate(x_test,y_test)
```

Key Libraries Used:

Model Architecture (Sequential Neural Network):

The project utilized a keras.models.Sequential model, which is a linear stack of layers. The architecture is as follows:

Input Layer (keras.Input(shape=(28, 28))) and Flatten Layer (keras.layers.Flatten()):

The input layer defines the shape of the input data, which are 28x28 pixel grayscale images.
The Flatten layer transforms the 28x28 2D image array into a 1D array of 784 pixels. This is necessary because the subsequent Dense layers expect a 1D input.
First Hidden Layer (keras.layers.Dense(300, activation="relu")):

A fully connected (Dense) layer with 300 neurons.
It uses the ReLU (Rectified Linear Unit) activation function. ReLU outputs the input directly if it's positive, otherwise, it outputs zero. This introduces non-linearity, allowing the model to learn complex patterns, and is computationally efficient.
Second Hidden Layer (keras.layers.Dense(100, activation="relu")):

Another fully connected (Dense) layer with 100 neurons.
Again, it uses the ReLU activation function for the same reasons as the first hidden layer.
Output Layer (keras.layers.Dense(10, activation="softmax")):

A fully connected (Dense) layer with 10 neurons, corresponding to the 10 different classes in the Fashion MNIST dataset (e.g., 'T-shirt/top', 'Trouser', 'Ankle boot').
It uses the Softmax activation function. Softmax converts the raw output scores (logits) into a probability distribution, where each value represents the probability of the input belonging to a specific class. The sum of all probabilities for an input will be 1, making it suitable for multi-class classification.

Environment Setup & Library Imports: The initial step involved setting up the Python environment by importing essential libraries like numpy for numerical operations, pandas for data handling, matplotlib.pyplot for visualization, and tensorflow with keras for building and training the neural network.

Model Definition and Architecture: A deep neural network model was constructed using Keras's Sequential API:

Reproducibility: Random seeds were set to ensure that the model's initialization and subsequent training process could be replicated.
Layer Construction: The model consisted of:

Two Dense (fully connected) hidden layers with ReLU activation functions, which introduce non-linearity and allow the model to learn complex patterns.
A final Dense output layer with 10 neurons (one for each class) and a softmax activation function, which outputs probabilities for each class, summing up to 1.
Inspection: The model's architecture was reviewed using model.summary() and visualized with keras.utils.plot_model().

Evaluation and Prediction: After training, the model's performance was assessed, and predictions were made:
